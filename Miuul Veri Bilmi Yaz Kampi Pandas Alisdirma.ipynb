```python
# Görev6:  embarked değişkeninin tipini kontrol ediniz. Tipini category olarakd eğiştiriniz ve tekrar kontrolediniz.
df.embarked.dtype
df.embarked=df.embarked.astype('category')
df.embarked.dtype
```

```python
# Görev7:  embarked değeriC olanlarıntümbilgelerinigösteriniz.
df[df['embarked']=='C']
```

```python
# Görev8:  embarked değeriS olmayanlarıntümbilgelerinigösteriniz.
df[df['embarked'] !='S']
```

```python
# Görev9:   Yaşı 30 dan küçük ve kadın olan yolcuların tümbilgilerini gösteriniz.
df[(df['age'] < 30) & (df['sex']=='female')]
```

```python
# Görev10:  Fare'i 500'den büyük ve ya yaşı 70 den büyük yolcuların bilgilerini gösteriniz.
df.loc[(df['fare'] > 500) & (df['age'] > 70)]
```

```python
# Görev 11:  Her birdeğişkendekiboşdeğerlerintoplamınıbulunuz.
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
# Görev 12:  who değişkeninidataframe’dençıkarınız.
df.drop('who', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Görev13:  deck değikenindeki boşdeğerleri deck değişkenin en çok tekrareden değeri(mode) ile doldurunuz
df['deck'].fillna(df['deck'].mode().values[0], inplace=True)
```

```python
# Görev14:  age değikenindekiboşdeğerleriage değişkeninmedyanıiledoldurunuz.
df['age'].fillna(df['age'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Görev15:  survived değişkeninin pclass ve cinsiyet değişkenleri kırılımınında sum, count, mean değerlerini bulunuz.
df.groupby(['pclass','sex']).agg({'survived':['sum','count','mean']})
```

```python
# Tekrar eden deyerler bakmak
df.duplicated().sum()
```

```python
# Tekrar eden deyerleri silmek
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
# Görev 16:  30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 vericek bir fonksiyon yazın. Yazdığınız fonksiyonukullanaraktitanikveri setinde age_flag adında birdeğişken oluşturunuzoluşturunuz. (apply velambda yapılarınıkullanınız)
df['age_flag']= df['age'].apply(lambda x:1 if x < 30 else 0)
```
